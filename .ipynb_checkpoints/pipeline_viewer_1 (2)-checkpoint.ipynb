```python
# Read JSON
with open('riscv_pipeline_1.txt', 'r') as f:
    json_lines = f.readlines()

inst_jsons = []
for line in json_lines:
    try:
        inst_jsons.append(json.loads(line))
    except json.JSONDecodeError:
        pass
```

```python
inst_ids = np.array([inst_jsons[i]["id"] for i in range(len(inst_jsons))])
inst_cycle = np.array([inst_jsons[i]["cycle"].strip() for i in range(len(inst_jsons))])
inst_event = np.array([inst_jsons[i]["event_name"] for i in range(len(inst_jsons))])
pc = np.array([inst_jsons[i]["data"] for i in range(len(inst_jsons))])
inst_parent = np.array([inst_jsons[i]["parents"] for i in range(len(inst_jsons))])
data = np.column_stack((inst_ids,inst_parent, inst_cycle, inst_event, pc))
columns = ["Instruction ID", "Parent ID", "Cycle", "Stage", "PC"]
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def get_all_instruction_threads(df):
    id = 0
    fetched_instructions = df[df["Stage"] == "IF1"]
    instruction_sequences = []
    for row in fetched_instructions.itertuples():
        pc = row.PC
        inst_trace = trace_instruction(df, row, [pc], [])
        instruction_sequences.extend(inst_trace)
    for row in instruction_sequences:
        row[0] = (id, row[0])
        id += 1
        if row[-1][0] == 'WB':
            row.append(('RET', str(int(row[-1][1]) + 1), None))
        else:
            row.append(('FLUSH', str(int(row[-1][1]) + 1), None))
    return instruction_sequences

def trace_instruction(df, row, curr_path, paths):
    stage = row.Stage
    cycle = row.Cycle
    pc = row.PC
    inst_id = row._1
    children = df[df["Parent ID"] == inst_id].drop_duplicates("PC")
    curr_path.append((stage, cycle, pc))
    if (children.shape[0] == 0):
        paths.append(curr_path)
        return paths
    else:
        for child in children.itertuples():
            paths.extend(trace_instruction(df, child, curr_path[:], []))
        return paths
```

```python
threads = get_all_instruction_threads(df.tail(20000))
```

```python
import heapq

def convert_to_kanata(threads, verbose=False):
    pq = []
    id = 0
    if not verbose:
        threads = list(filter(lambda x: x[-1][0] == 'RET', threads)) #Relies on the last element of inst list being RET
    for inst in threads:
        for stage in inst[1:]:
            heapq.heappush(pq, ((int(stage[1])), (id, stage[2], stage[0]))) #Min heap of (cycle -> (unique_id, pc, pipeline stage))
        id += 1
            
    with open('konata.txt', 'w') as file:
        file.write('Kanata	0004\n')
        cycle, (id, pc, stage) = heapq.heappop(pq)
        prev_cycle = cycle
        file.write(f'C=	{cycle}\n')
        while pq:
            cycle_diff = cycle - prev_cycle
            print(cycle_diff)
            if (cycle_diff > 0):
                file.write(f"C	{cycle_diff}\n")
            if (stage == 'IF1'):
                file.write(f"I	{id}	0	0\n")
                # file.write(f"L	{id}	0	{pc}\n")
            if (stage == 'WB'):
                file.write(f"L	{id}	0	PC: {pc}\n")
            if (stage == 'RET'):
                file.write(f"R	{id}	{id}	0\n")
            elif (stage == 'FLUSH'):
                file.write(f"R	{id}	{id}	1\n")
            else:
                file.write(f"S	{id}	0	{stage}\n")
                file.write(f"L	{id}	1	PC: {pc}\n")

            prev_cycle = cycle
            cycle, (id, pc, stage) = heapq.heappop(pq)



        
        


```
